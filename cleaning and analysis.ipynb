```python
customers_geo = df_customer.merge(
    df_geolocation.drop_duplicates(subset=["geolocation_zip_code_prefix"]),
    left_on="customer_zip_code_prefix",
    right_on="geolocation_zip_code_prefix",
    how="left"
)

# Merge customers+geo with orders
df_master = df_orders.merge(customers_geo, on="customer_id", how="left")

# Merge with order_items
df_master = df_master.merge(df_order_items, on="order_id", how="left")

# Merge with products
df_master = df_master.merge(df_products, on="product_id", how="left")

# Merge with sellers
df_master = df_master.merge(df_sellers, on="seller_id", how="left")

# Merge with order_payments
df_master = df_master.merge(df_order_payments, on="order_id", how="left")

# Merge with order_reviews
df_master = df_master.merge(df_order_reviews, on="order_id", how="left")
```

```python
df_master = df_master.merge(df_product_category, on='product_category_name', how='left')
```

```python
date_columns = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date',
    'shipping_limit_date',
    'review_creation_date',
    'review_answer_timestamp'
]

for col in date_columns:
    df_master[col] = pd.to_datetime(df_master[col], errors='coerce')  # coerce converts invalid values to NaT
```

```python
df_master.isnull().sum()
```

```python
df_master.duplicated().sum()
```

```python
# Drop rows with missing key identifiers
df_master = df_master.dropna(subset=['order_id', 'customer_id', 'product_id', 'seller_id', 'price', 'freight_value'])
```

```python
# Fill product categories with 'other'
df_master['product_category_name'] = df_master['product_category_name'].fillna('other')
df_master['product_category_name_english'] = df_master['product_category_name_english'].fillna('other')
```

```python
# Fill payment_type missing with None (null)
df_master['payment_type'] = df_master['payment_type'].where(df_master['payment_type'].notna(), None)
```

```python
# Geolocation: Fill from zip_code_prefix groups
df_master[['geolocation_lat', 'geolocation_lng']] = df_master.groupby('geolocation_zip_code_prefix')[['geolocation_lat', 'geolocation_lng']].transform(lambda x: x.fillna(x.mean()))

# Leave meaningful nulls as they are (delivery dates, reviews)
```

```python
df_master.dropna(subset=['payment_sequential'], inplace=True)
```

```python
df_master.info()
```

```python
# Calculate delivery time in days
df_master['order_delivery_time'] = (df_master['order_delivered_customer_date'] - df_master['order_purchase_timestamp']).dt.days
```

**Sales & Revenue Insights**

```python
# 1. Total revenue & orders
total_revenue = df_master['price'].sum()
total_orders = df_master['order_id'].nunique()
print('Total Revenue:', total_revenue)
print('Total Orders:', total_orders)
```

```python
# 2. Revenue by month
df_master['purchase_month'] = df_master['order_purchase_timestamp'].dt.to_period('M')
monthly_revenue = df_master.groupby('purchase_month')['price'].sum().reset_index()
print(monthly_revenue)
```

```python
# 3. Average order value
avg_order_value = total_revenue / total_orders
print('Avg Order Value:', avg_order_value)
```

**Delivery & Fulfillment Performance**

```python
# Average delivery time
avg_delivery_time = df_master['order_delivery_time'].mean()

# Late deliveries
df_master['late_delivery'] = df_master['order_delivered_customer_date'] > df_master['order_estimated_delivery_date']
late_delivery_rate = df_master['late_delivery'].mean() * 100

print('Avg Delivery Time:', avg_delivery_time)
print('Late Delivery Rate:', late_delivery_rate)
```

**Customer & Product Insights**

```python
# Top selling categories
top_categories = df_master.groupby('product_category_name_english')['order_id'].count().sort_values(ascending=False).head(10)

# Revenue by category
category_revenue = df_master.groupby('product_category_name_english')['price'].sum().sort_values(ascending=False).head(10)

# Review score distribution
review_distribution = df_master['review_score'].value_counts(normalize=True) * 100

print('---Top Categories---\n',top_categories)
print('\n---Category Revenue---\n', category_revenue)
print('\n---Review Distribution---\n', review_distribution)
```
